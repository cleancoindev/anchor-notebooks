# Transaction history

```python
from requests import get
from json import dumps

wallet_address = 'terra12hnhh5vtyg5juqnzm43970nh4fw42pt27nw9g9'
# endpoint = 'https://api.anchorprotocol.com/api/v1' # is mainnet
endpoint = 'https://tequila-api.anchorprotocol.com/api/v1'

result = get(endpoint + '/history/' + wallet_address)
result_data = result.json()

print('next', result_data['next'])
print('history', dumps(result_data['history'], indent=2))

# get next page
next_result = get(endpoint + '/history/' + wallet_address + '?offset=' + result_data['next'])
next_result_data = next_result.json()

print('next', dumps(next_result_data, indent=2))
```
